```python
import re

contractions_dict = {    
    "ain't": "am not",
    "aren't": "are not",
    "can't": "cannot",
    "can't've": "cannot have",
    "'cause": "because",
    "could've": "could have",
    "couldn't": "could not",
    "couldn't've": "could not have",
    "didn't": "did not",
    "doesn't": "does not",
    "don't": "do not",
    "hadn't": "had not",
    "hadn't've": "had not have",
    "hasn't": "has not",
    "haven't": "have not",
    "he'd": "he had",
    "he'd've": "he would have",
    "he'll": "he will",
    "he'll've": "he will have",
    "he's": "he is",
    "how'd": "how did",
    "how'd'y": "how do you",
    "how'll": "how will",
    "how's": "how is",
    "i'd": "i had",
    "i'd've": "i would have",
    "i'll": "i will",
    "i'll've": "i will have",
    "i'm": "i am",
    "i've": "i have",
    "isn't": "is not",
    "it'd": "it had",
    "it'd've": "it would have",
    "it'll": "it will",
    "it'll've": "iit will have",
    "it's": "it is",
    "let's": "let us",
    "ma'am": "madam",
    "mayn't": "may not",
    "might've": "might have",
    "mightn't": "might not",
    "mightn't've": "might not have",
    "must've": "must have",
    "mustn't": "must not",
    "mustn't've": "must not have",
    "needn't": "need not",
    "needn't've": "need not have",
    "o'clock": "of the clock",
    "oughtn't": "ought not",
    "oughtn't've": "ought not have",
    "shan't": "shall not",
    "sha'n't": "shall not",
    "shan't've": "shall not have",
    "she'd": "she had",
    "she'd've": "she would have",
    "she'll": "she will",
    "she'll've": "she will have",
    "she's": "she is",
    "should've": "should have",
    "shouldn't": "should not",
    "shouldn't've": "should not have",
    "so've": "so have",
    "so's": "so is",
    "that'd": "that had",
    "that'd've": "that would have",
    "that's": "that is",
    "there'd": "there had",
    "there'd've": "there would have",
    "there's": "there is",
    "they'd": "they had",
    "they'd've": "they would have",
    "they'll": "they will",
    "they'll've": "they will have",
    "they're": "they are",
    "they've": "they have",
    "to've": "to have",
    "wasn't": "was not",
    "we'd": "we had",
    "we'd've": "we would have",
    "we'll": "we will",
    "we'll've": "we will have",
    "we're": "we are",
    "we've": "we have",
    "weren't": "were not",
    "what'll": "what will",
    "what'll've": "what will have",
    "what're": "what are",
    "what's": "what is",
    "what've": "what have",
    "when's": "when is",
    "when've": "when have",
    "where'd": "where did",
    "where's": "where is",
    "where've": "where have",
    "who'll": "who will",
    "who'll've": "who will have",
    "who's": "who is",
    "who've": "who have",
    "why's": "why is",
    "why've": "why have",
    "will've": "will have",
    "won't": "will not",
    "won't've": "will not have",
    "would've": "would have",
    "wouldn't": "would not",
    "wouldn't've": "would not have",
    "y'all": "you all",
    "y'all'd": "you all would",
    "y'all'd've": "you all would have",
    "y'all're": "you all are",
    "y'all've": "you all have",
    "you'd": "you had",
    "you'd've": "you would have",
    "you'll": "you will",
    "you'll've": "you will have",
    "you're": "you are",
    "you've": "you have"
}
 
 
def expand_contractions(text, contractions_dict):
    contractions_pattern = re.compile('({})'.format('|'.join(contractions_dict.keys())),
                                      flags=re.IGNORECASE | re.DOTALL)
                                    # re.IGNORECASE: case insensitive
                                    # re.DOTALL: titik dianggap titik, bukan sebagai wildcard
 
    def expand_match(contraction):
        match = contraction.group(0)
        first_char = match[0]
        expanded_contraction = contractions_dict.get(match) \
            if contractions_dict.get(match) \
            else contractions_dict.get(match.lower())
        expanded_contraction = expanded_contraction
        return expanded_contraction
 
    expanded_text = contractions_pattern.sub(expand_match, text)    # dari singkatan, menjadi dijabarkan
    expanded_text = re.sub("'", "", expanded_text)                  # menghilangkan petik satu
    return expanded_text
 
def main_contraction(text):
    text = expand_contractions(text, contractions_dict)
    return text
```

```python
from textblob import TextBlob

exclude_words = {'kimi', 'andrea', 'antonelli', 'mimi'}

def correct_spelling_exclude(text):
    corrected_words = []
    for word in text.split():
        # Check if the word is in the exclusion list
        if word.lower() in exclude_words:
            corrected_words.append(word)  # Keep the word as is
        else:
            corrected_word = str(TextBlob(word).correct())  # Apply spell correction
            corrected_words.append(corrected_word)
    
    return ' '.join(corrected_words)
```

```python
def remove_whitespaces(text):
    return ' '.join(text.split())
```

```python
def remove_single_characters(text):
    return ' '.join([word for word in text.split() if len(word) > 1])
```

```python
def remove_stopwords(text):
    # Refine stopwords removal to exclude certain important words
    return ' '.join([word for word in text.split() if word not in english_stopwords])
```

```python
def kamus(corpus):
    # Extract all alphabetic words from the corpus
    text = corpus.str.extractall('([A-Za-z]+)')
    text.columns = ['check']
    
    # Reset index for clean DataFrame
    b = text.reset_index(drop=True)
    
    # Filter out single-character words
    b = b[b['check'].str.len() > 1]
    
    # Count the frequency of each word
    text = b['check'].value_counts()
    
    # Create a DataFrame to store the word ('kata') and its frequency ('freq')
    kamus = pd.DataFrame({
        'kata': text.index,
        'freq': text.values
    })

    return kamus
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Manually set the NLTK data path to include your specific path
nltk.data.path.append('/Users/anthoniusbimo/nltk_data')

# Initialize the SentimentIntensityAnalyzer with the full path to vader_lexicon.txt
sentiments = SentimentIntensityAnalyzer(lexicon_file='/Users/anthoniusbimo/nltk_data/sentiment/vader_lexicon/vader_lexicon.txt')

# Calculate sentiment scores for each text
df2['Positive'] = [sentiments.polarity_scores(i)['pos'] for i in df2['text_remove_stopwords_lemma']]
df2['Neutral'] = [sentiments.polarity_scores(i)['neu'] for i in df2['text_remove_stopwords_lemma']]
df2['Negative'] = [sentiments.polarity_scores(i)['neg'] for i in df2['text_remove_stopwords_lemma']]
df2['Compound'] = [sentiments.polarity_scores(i)['compound'] for i in df2['text_remove_stopwords_lemma']]

# Display the DataFrame
df2.head()
```

```python
# Function to label sentiment based on polarity score
def labelling(value):
    if value > 0:
        label = "Positive"
    elif value < 0:
        label = "Negative"
    else:
        label = "Neutral"
    return label

# Apply TextBlob and VADER sentiment labeling to DataFrame
df2['TextBlob Label'] = df2['polarity'].apply(labelling)
df2['Vader Label'] = df2['Compound'].apply(labelling)

# Display the updated DataFrame with labels
df2.head()
```

```python
df_sentiment = df2.pivot_table(
    index='TextBlob Label',  # Rows: Sentiment labels from TextBlob
    columns='Vader Label',   # Columns: Sentiment labels from VADER
    values='text_remove_stopwords_lemma',  # Values: Text data column
    aggfunc='count'  # Aggregation function: Count the number of comments
)
```

```python
from langdetect import detect 
import re
import pandas as pd

# Function to detect language
def detect_language(text):
    try:
        return detect(str(text))
    except:
        return None  # Return None if detection fails

# Function to check if the text contains non-Latin characters
def contains_non_latin(text):
    latin_pattern = re.compile(r'^[a-zA-Z0-9\s.,!?]*$')  # Define a pattern that matches only Latin characters
    return not bool(latin_pattern.match(text))  # Return True if non-Latin characters are found, False otherwise

# Apply functions to your DataFrame
df2['Detected_Language'] = df2['text_remove_stopwords_lemma'].apply(detect_language)
df2['Contains_Non_Latin'] = df2['text_remove_stopwords_lemma'].apply(contains_non_latin)
```
